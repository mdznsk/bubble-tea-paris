```python
import scrapy
from scrapy.crawler import CrawlerProcess
import os
import logging
```

```python
class YelpSpider(scrapy.Spider):
    name = "yelp"
    start_urls = ['https://www.yelp.fr/']

    def parse (self, response):
        return scrapy.FormRequest.from_response(
            response,
            formdata={'find_desc': 'bubble tea', 'find_loc': 'paris'},
            callback=self.after_search
        )

    def after_search(self, response):
        results = response.css('h4 a')
        for r in results:
            yield { 
                'name': r.css('::text').get()}
        try:
            next = response.css('a.next-link').attrib["href"]
        except KeyError:
            logging.info('No next page. End crawl.')
        else:
            yield response.follow(next, callback=self.after_search)
```

```python
filename = 'yelp-names.json'
if filename in os.listdir('/Users/eline/Documents/bb tea project/'):
        os.remove('/Users/eline/Documents/bb tea project/' + filename)

# Declare a new CrawlerProcess with some settings
process = CrawlerProcess(settings = {
    'USER_AGENT': 'Mozilla/4.0 (compatible; MSIE 7.0; Windows NT 5.1)',
    'LOG_LEVEL': logging.INFO,
    "FEEDS": {
        '/Users/eline/Documents/bb tea project/' + filename : {"format": "json"},
    }
})

# Start the crawling using the spider you defined above
process.crawl(YelpSpider)
process.start()
```
